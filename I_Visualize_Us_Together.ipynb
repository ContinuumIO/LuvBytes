```python
import bokeh, bokeh.plotting, IPython, numpy, scipy.optimize
from coffeetools import coffee
bokeh.plotting.output_notebook(resources=bokeh.resources.CDN)
```

```python
N = 750

xx = [-.7,.7]; yy = [.2,.2];
p = bokeh.plotting.figure( tools=['pan','xwheel_zoom'], title = 'Please, help bring us closer together')
p.circle( x=xx,y=yy,size=50, fill_alpha=.6)
p.text( x=xx[0],y=yy[0],text=['S'], text_align='left')
p.text( x=xx[1],y=yy[1],text=['U'], text_align='right')

tt = numpy.linspace(0,2*numpy.pi,N)
xx = (10*numpy.sin(tt)**3)/10
yy = (13*numpy.cos(tt)-5*numpy.cos(2*tt)-2**numpy.cos(3*tt)-numpy.cos(4*tt))/13
mult = numpy.random.rand(N) 
xx = numpy.multiply(xx,mult)
yy = numpy.multiply(yy,mult)

p.circle( x=xx,y=yy,radius = .1,fill_color=['rgba('+','.join([str(v) for v in [int(numpy.random.rand()*255) for j in range(3)]])+',.4)' for i in range(N)], line_alpha=0)
p.x_range.callback = bokeh.models.CustomJS(
    args = {
        'fig': p,
        'xx': p.x_range,
    },
    code = coffee.compile("""
    dx = xx.get('end') - xx.get('start')
    switch
        when dx < 20
            fig.set 'title', 'Please, help bring us closer together'
        when dx <  40
            fig.set 'title', 'Thanks for your help.'
        when dx < 60
            fig.set 'title', 'We are better when we work together.'
        when dx < 90
            fig.set 'title', 'Almost!'
        when dx < 6000
            fig.set 'title', 'Together at last'
        else
            console.log 
            fig.set 'title', ['forever','and ever..'][Math.floor(Math.log10(dx))%%2]"""))
IPython.display.display(IPython.display.Markdown("# I Visualize Us Together"),IPython.display.HTML(''.join(bokeh.embed.components(p))))
```
